***Everything*** in pytorch is a tensor but for standard convention we name tensor with zero dimension i.e having only one element as scalar
and tensor with only one dimension as vector, note that a vector one dimension means that it has only one column.
A tensor with two dimension is called a matrix, it has more than one column,
also any tensor with more than two dimension is simply called as a tensor.
a pair of square bracket represents a dimension.
if you perform operation on two tensor they must have same dimensions
i.e they must have the same dimension as well as same number of column and shape.
to understand shape you must visualise,[2,2,3]->it means two rows and two columns , there are two layers , 3 elements in each axis.

```python
# creating tensors
# most most used 
# visualise it is a 3d cube
TENSOR = torch.tensor([[[7,7],[6,6]],[[0,0],[5,3]]])
print(TENSOR)
print(TENSOR.shape)
print(TENSOR.ndim)
```

### MATRIX MULTIPLICATION AND ELEMENT WISE MULTIPLICATION

```python
tensor1 = torch.rand(5,5)
tensor2 = torch.rand(5,5)
tensor1 * tensor2
## element wise multiplication
## shape dimension must be same
```

```python
## matrix multiplication
%%time
tensor1 = torch.rand(3,5)
tensor2 =  torch.rand(5,3)
torch.matmul(tensor1,tensor2)
```

```python
tensor1 = torch.rand(3,2)
tensor2 = torch.rand(3,2)
## you cant multiply them but  you can if you take  transpose of one of them
torch.mm(tensor1 ,tensor2.T)
```

```python
### tensor aggregation
x = torch.arange(0,11,1)
print(x)
print(torch.min(x))
print(torch.max(x))
print(torch.mean(x.type(torch.float32)))   ## you have to specify the output of the mean, its special..
print(torch.mean(x.type(torch.float16)))   ## you ave to specify float16/float32..
print(torch.sum(x))
```

### ***RESHAPING, STACKING , UNSQUEEZING TENSORS***


```python
### view used for reshaping
x = torch.rand(3,4)
print(x)
print(x.view(12))
print(x.view(4,3))
print(x)
```

```python
### reshape used for reshaping
x = torch.rand(3,4)
print(x)
print(x.reshape(12))
print(x.reshape(4,3))
print(x)
```

### difference between reshape and view
* reshape reshapes an input tensor to a defined shape
* view returns a view of the input tensor with the defined shape but keeps the same memory as the original tensor and hence changing the view will change the original tensor

```python
tensor = torch.rand(3,4)
print(tensor)
tensor.view(4,3)+10
print(tensor)
```
